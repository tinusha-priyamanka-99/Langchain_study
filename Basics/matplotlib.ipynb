```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x, y, c='g')
plt.title('Scatter Plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
y = x*x
```

```python
plt.plot(x,y,'y*',linestyle = 'dashed', markersize=15)
```
