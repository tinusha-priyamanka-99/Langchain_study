```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x, y, c='g')
plt.title('Scatter Plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
y = x*x
```

```python
plt.plot(x,y,'y*',linestyle = 'dashed', markersize=15)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r.')
plt.subplot(2,2,2)
plt.plot(x,y,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'b-')
```

```python
x= np.arange(1,15)
y = 3*x+20
plt.title('Demo')
plt.plot(x,y,)
```

```python
np.pi
```

```python
x = np.arange(0,4*np.pi, 0.1)
y = np.sin(x)
plt.plot(x,y)
```

```python
x = np.arange(0,5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine')
```

```python
x = [2,8,10]
y = [11,16,9]

x2= [3,7,9]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2, color='g')
plt.title('Bar graph')
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins=10) #default
plt.title("Histogram")
```

```python
data = [np.random.normal(0,std, 100) for std in range(1,4)]

plt.boxplot(data, vert=True, patch_artist=True)
```

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True)
```
